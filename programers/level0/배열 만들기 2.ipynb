```python
def solution(l, r):
    answer = []
    for num in range(l, r+1):
        if num%5==0:
            is_in = True
            for n in str(num):
                if n not in ['0','5']:
                    is_in = False
                    
                    break

            if is_in:
                answer.append(num)
            

    return answer if answer else [-1]
```

```python
solution(5,555)
```
