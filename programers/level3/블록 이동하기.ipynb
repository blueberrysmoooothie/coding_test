```python
from queue import Queue
def solution(board):
    N= len(board)
    board = [[1]*(N+2)] + [[1]+B +[1] for B in board] + [[1]*(N+2)]

    visited = {(r,c):[0,0] for r in range(N+2) for c in range(N+2)}
    to_visit = Queue()
    to_visit.put([1,1, 0, 0])

    while True:
        r1, c1, vertical, cnt = to_visit.get()
        r2, c2 = r1+vertical , c1+1- vertical
        if r2+c2 == 2*N:
            return cnt
        # 가로 
        if not vertical:
            visited[(r1,c1)][0]=1
            # 횡 이동
            if not visited[(r1,c1-1)][0]:
                if not board[r1][c1-1]:
                    to_visit.put([r1,c1-1,0, cnt+1])
            
            if not visited[(r2,c2)][0]:
                if not board[r2][c2+1]:
                    to_visit.put([r2,c2,0, cnt+1])

            if board[r1-1][c1:c1+2]==[0,0]:
                # 종 이동
                if not visited[(r1-1,c1)][0]:
                    to_visit.put([r1-1,c1,0, cnt+1])

                # 회전
                if not visited[(r1-1,c1)][1]:
                    to_visit.put([r1-1,c1,1, cnt+1])
                if not visited[(r2-1,c2)][1]:
                    to_visit.put([r2-1,c2,1, cnt+1])

            if board[r1+1][c1:c1+2]==[0,0]:
                if not visited[(r1+1,c1)][0]:
                    to_visit.put([r1+1,c1,0, cnt+1])
                if not visited[(r1,c1)][1]:
                    to_visit.put([r1,c1,1, cnt+1])
                if not visited[(r2,c2)][1]:
                    to_visit.put([r2,c2,1, cnt+1])

        # 세로
        else:
            visited[(r1,c1)][1]=1
            # 종 이동
            if not visited[(r1-1,c1)][1]:
                if not board[r1-1][c1]:
                    to_visit.put([r1-1,c1,1, cnt+1])
            
            if  not visited[(r2,c2)][1]:
                if not board[r2+1][c2]:
                    to_visit.put([r2,c2,1, cnt+1])


            if board[r1][c1-1]==0 and  board[r2][c2-1] == 0:
                # 종 이동
                if  not visited[(r1,c1-1)][1]:
                    to_visit.put([r1,c1-1,1, cnt+1])

                # 회전
                if not visited[(r1,c1-1)][0]:
                    to_visit.put([r1,c1-1,0, cnt+1])
                if not  visited[(r2,c2-1)][0]:
                    to_visit.put([r2,c2-1,0, cnt+1])

            if board[r1][c1+1]==0 and board[r2][c2+1]==0:
                if  not visited[(r1,c1+1)][1]:
                    to_visit.put([r1,c1+1,1, cnt+1])
                if  not visited[(r1,c1)][0]:
                    to_visit.put([r1,c1,0, cnt+1])
                if  not visited[(r2,c2)][0]:
                    to_visit.put([r2,c2,0, cnt+1])
    
```

```python
class Node:
    def __init__(self):
        self.visited = [False, False]

class Map_:
    def __init__(self, board):
        self.N = len(board)
        board = [[1]*(self.N+2)] + [[1]+B+[1] for B in board] + [[1]*(self.N+2)]
        self.board = [[None]*(self.N+2)]
        for l in range(1, self.N+1):
            self.board.append([])
            for c in range(self.N+2):
                if not board[l][c] :
                    self.board[l].append(Node())
                else: 
                    self.board[l].append(None)
        self.board += [[None]*(self.N+2)]

    def bfs(self):
        to_visit = [[1,1, 0, 0]]
        for r1, c1, vertical, cnt in to_visit:
            if r1+c1 == self.N *2 -1:
                return cnt

            # 가로 
            if not vertical:
                self.board[r1][c1].visited[0]=True
                # 횡 이동
                if self.board[r1][c1-1] is not None:
                    if not self.board[r1][c1-1].visited[0]:
                        to_visit.append([r1,c1-1,0, cnt+1])
                
                if self.board[r1][c1+2] is not None:
                    if not self.board[r1][c1+1].visited[0]:
                        to_visit.append([r1,c1+1,0, cnt+1])

                if self.board[r1-1][c1]is not None and self.board[r1-1][c1+1] is not None:
                    # 종 이동
                    if not self.board[r1-1][c1].visited[0]:
                        to_visit.append([r1-1,c1,0, cnt+1])

                    # 회전
                    if not self.board[r1-1][c1].visited[1]:
                        to_visit.append([r1-1,c1,1, cnt+1])
                    if not self.board[r1-1][c1+1].visited[1]:
                        to_visit.append([r1-1,c1+1,1, cnt+1])

                if self.board[r1+1][c1]is not None and self.board[r1+1][c1+1] is not None:
                    if not self.board[r1+1][c1].visited[0]:
                        to_visit.append([r1+1,c1,0, cnt+1])
                    if not self.board[r1][c1].visited[1]:
                        to_visit.append([r1,c1,1, cnt+1])
                    if not self.board[r1][c1+1].visited[1]:
                        to_visit.append([r1,c1+1,1, cnt+1])

            # 세로
            else:
                self.board[r1][c1].visited[1]=True
                # 종 이동
                if self.board[r1-1][c1] is not None:
                    if not self.board[r1-1][c1].visited[1]:
                        to_visit.append([r1-1,c1,1, cnt+1])
                
                if self.board[r1+2][c1]is not None:
                    if not self.board[r1+1][c1].visited[1]:
                        to_visit.append([r1+1,c1,1, cnt+1])


                if self.board[r1][c1-1]is not None and self.board[r1+1][c1-1] is not None:
                    # 횡 이동
                    if not self.board[r1][c1-1].visited[1]:
                        to_visit.append([r1,c1-1,1, cnt+1])

                    # 회전
                    if not self.board[r1][c1-1].visited[0]:
                        to_visit.append([r1,c1-1,0, cnt+1])
                    if not self.board[r1+1][c1-1].visited[0]:
                        to_visit.append([r1+1,c1-1,0, cnt+1])

                if self.board[r1][c1+1]is not None and self.board[r1+1][c1+1] is not None:
                    if not self.board[r1][c1+1].visited[1]:
                        to_visit.append([r1,c1+1,1, cnt+1])
                    if not self.board[r1][c1].visited[0]:
                        to_visit.append([r1,c1,0, cnt+1])
                    if not self.board[r1+1][c1].visited[0]:
                        to_visit.append([r1+1,c1,0, cnt+1])
    

def solution(board):
    m = Map_(board)
    return m.bfs()
    
```

```python
from queue import Queue

class Map_:
    def __init__(self, board):
        self.N = len(board)
        self.board = [[None]*(self.N+2)]
        for l in range(1, self.N+1):
            self.board.append([None])
            for c in range(1,self.N+1):
                if not board[l-1][c-1] :
                    self.board[l].append(1)
                else: 
                    self.board[l].append(None)
            self.board[l].append(None)
        self.board += [[None]*(self.N+2)]

    def can_move(self, cur1, cur2):
        cand=[]
        deltas = [(-1,0), (1,0), (0,-1), (0,1)]
        # 직진
        for dr, dc in deltas:
            n_r1, n_c1 = cur1[0]+dr, cur1[1]+dc
            n_r2, n_c2 = cur2[0]+dr, cur2[1]+dc
            if self.board[n_r1][n_c1] is not None and self.board[n_r2][n_c2] is not None:
                cand.append([(n_r1, n_c1), (n_r2, n_c2)])

        # 회전
        if cur1[0]==cur2[0]:

            d= -1
            if self.board[cur1[0]+d][cur1[1]] is not None and self.board[cur2[0]+d][cur2[1]] is not None:
                    cand.append([(cur1[0]+d, cur1[1]), cur1])
                    cand.append([(cur2[0]+d, cur2[1]), cur2])
            d= 1
            if self.board[cur1[0]+d][cur1[1]] is not None and self.board[cur2[0]+d][cur2[1]] is not None:
                    cand.append([cur1, (cur1[0]+d, cur1[1])])
                    cand.append([cur2, (cur2[0]+d, cur2[1])])
                    
        else:
            d= -1
            if self.board[cur1[0]][cur1[1]+d] is not None and self.board[cur2[0]][cur2[1]+d] is not None:
                    cand.append([(cur1[0], cur1[1]+d),cur1])
                    cand.append([(cur2[0], cur2[1]+d),cur2])
            d= 1
            if self.board[cur1[0]][cur1[1]+d] is not None and self.board[cur2[0]][cur2[1]+d] is not None:
                    cand.append([cur1, (cur1[0], cur1[1]+d)])
                    cand.append([cur2, (cur2[0], cur2[1]+d)])
        return cand
        
    def bfs(self):
        to_visit = Queue()
        to_visit.put([(1,1), (1, 2), 0])
        check = set([((1,1), (1, 2))])
        while to_visit:
            cur1, cur2, cnt = to_visit.get()
            # print([cur1, cur2, cnt], self.can_move(cur1, cur2))
            if (self.N, self.N) == cur2:
                return cnt
            for nl, nc in self.can_move(cur1, cur2):
                if (nl, nc) not in check:
                    to_visit.put([nl, nc, cnt+1])
                    check.add((nl,nc))
            

def solution(board):
    m = Map_(board)
    return m.bfs()
    
```
