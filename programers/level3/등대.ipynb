```python
import sys
sys.setrecursionlimit(10**6)
class node:
    def __init__(self):
        self.on = False
        self.edges = []

class Net:
    def __init__(self, n, lighthouse):
        self.nodes = [node() for _ in range(n+1)]
        for light in lighthouse:
            self.nodes[light[0]].edges.append(light[1])
            self.nodes[light[1]].edges.append(light[0])
        self.node_e = {}
        for i in range(1,n+1):
            e = len(self.nodes[i].edges)
            if e in self.node_e:
                self.node_e[e].append(i)
            else:
                self.node_e[e]=[i]

    def dfs(self, cur, visited=[]):
        nod = self.nodes[cur]
        for n in nod.edges:
            if len(self.nodes[n].edges) ==1:
                continue
            
            elif n not in visited:
                self.dfs(n,visited+[cur])

        if len(nod.edges) ==1:
                return 0
        for n in nod.edges:
            if not self.nodes[n].on:
                if n in visited:
                    continue
                nod.on = True
                break

    def on(self):
        k = sorted(list(self.node_e.keys()))[0]
        cur = self.node_e[k][0]
        self.dfs(cur)
        if not self.nodes[self.nodes[cur].edges[0]].on:
            return sum([n.on for n in self.nodes])+1
        
        return sum([n.on for n in self.nodes])
def solution(n, lighthouse):

    net = Net(n, lighthouse)


    answer = net.on()
    return answer

```

```python
import bisect
def solution(n, lighthouse):
    edges = {i:[] for i in range(1,n+1)}
    for a, b in lighthouse:
        edges[a].insert(bisect.bisect_left(edges[a],b),b)
        edges[b].insert(bisect.bisect_left(edges[b],a),a)

    ons = [0 for i in range(n)]
    while True:
        flg = True
        for i in range(1, n+1):
            if len(edges[i]) ==1:
                cur = edges[i][0]
                ons[cur-1] = 1
                for n in edges[cur]:
                    edges[n].pop(bisect.bisect_left(edges[n], cur))
                edges[cur] = []
                flg = False
        # print(ons)
        if flg:
            break
    return sum(ons)
```
