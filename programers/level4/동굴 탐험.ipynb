```python
class Node:
    def __init__(self):
        self.connection=[]
        self.before_visit = None
        self.after_visit = None

class Tree:
    def __init__(self,n, path, order) -> None:
        self.L = n
        self.tree = [Node() for _ in range(self.L)]
        for n1, n2 in path:
            self.tree[n1].connection.append(n2)
            self.tree[n2].connection.append(n1)

        for b, f in order:
            self.tree[b].after_visit = f
            self.tree[f].before_visit = b


    def bfs(self):
        queue = [0]
        cant_visit = {}
        visited = []
        while queue:
            cur = queue.pop(0)
            # print(cur,queue, cant_visit)

            if self.tree[cur].after_visit is not None:
                queue.append(self.tree[cur].after_visit)
                self.tree[self.tree[cur].after_visit].before_visit = None
                if cur in cant_visit:
                    del cant_visit[cur]

            for c in self.tree[cur].connection:
                if c in visited:
                    continue
                if self.tree[c].before_visit is not None:
                    cant_visit[self.tree[c].before_visit] = c
                else:
                    queue.append(c)

            visited.append(cur)
        if cant_visit:
            return False
        return True
        


def solution(n, path, order):
    tree = Tree(n,path, order)
    return tree.bfs()
```

```python
from collections import deque

class Node:
    def __init__(self):
        self.connection=[]

class Tree:
    def __init__(self,n, path, order) -> None:
        self.L = n
        self.tree = [Node() for _ in range(self.L)]
        for n1, n2 in path:
            self.tree[n1].connection.append(n2)
            self.tree[n2].connection.append(n1)

        self.before_visit = [0]*self.L
        self.after_visit = [0]*self.L
        for b, f in order:
            self.after_visit[b] = f
            self.before_visit[f] = b


    def bfs(self):
        queue = deque([0])
        cant_visit = {}
        visited = [0]*self.L
        while queue:
            cur = queue.popleft()
            if self.before_visit[cur]:
                cant_visit[self.before_visit[cur]] = cur
                continue

            if self.after_visit[cur]:
                self.before_visit[self.after_visit[cur]] = 0

                if cur in cant_visit:
                    queue.append(self.after_visit[cur])
                    del cant_visit[cur]

            for c in self.tree[cur].connection:
                if visited[c]:
                    continue
                else:
                    queue.append(c)

            visited[cur] = 1
            # print(cur,queue, cant_visit)
        if cant_visit:
            return False
        return True
        


def solution(n, path, order):
    tree = Tree(n,path, order)
    return tree.bfs()
```
