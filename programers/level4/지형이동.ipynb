```python
import numpy as np
import pandas as pd
def find_close_cell(land, visit_mat, height, height_type, f_i, f_j):

    find_stack = [[f_i,f_j]]
    for loc in find_stack:
        i, j = loc[0], loc[1]
        visit_mat[i][j] = height_type 
        
        if i!=0:
            if (visit_mat[i-1][j] == 0) and abs(land[i][j] - land[i-1][j])<=height:
                find_stack.append([i-1,j])
            
        if i!=(len(land)-1):
            if (visit_mat[i+1][j] == 0) and abs(land[i][j] - land[i+1][j])<=height:
                find_stack.append([i+1,j])
            
        if j!=0:
            if (visit_mat[i][j-1] == 0) and abs(land[i][j] - land[i][j-1])<=height:
                find_stack.append([i,j-1])
            
        if j!=(len(land)-1) :
            if (visit_mat[i][j+1] == 0) and abs(land[i][j] - land[i][j+1])<=height:
                find_stack.append([i,j+1])
    return visit_mat
    
    
def get_min_height_diff(land, visit_mat, group1, group2):
    min_height_diff = 10000
    for i in range(len(land)):
        for j in range(len(land)):
            if visit_mat[i][j]== group1:
                group2_cells = []
                
                if i !=0:
                    if (visit_mat[i-1][j] == group2):
                        group2_cells.append(land[i-1][j])
                
                if i !=(len(land)-1):
                    if (visit_mat[i+1][j] == group2):
                        group2_cells.append(land[i+1][j])
                
                if j !=0:
                    if (visit_mat[i][j-1] == group2):
                        group2_cells.append(land[i][j-1])
                
                if j !=(len(land)-1):
                    if (visit_mat[i][j+1] == group2):
                        group2_cells.append(land[i][j+1])
                
                if len(group2_cells):
                    for h in group2_cells:
                        if min_height_diff >= abs(land[i][j]-h):
                            min_height_diff = abs(land[i][j]-h)

    return min_height_diff
            
            

def solution(land, height):
    visit_mat = [[0] * len(land) for i in range(len(land))]
    height_type = 1
    for i in range(len(land)):
        for j in range(len(land)):        
            if visit_mat[i][j] == 0:
                visit_mat = find_close_cell(land, visit_mat, height, height_type, i, j)                
                height_type += 1

                
    print(visit_mat)
    
    answer = 0
    visited_group = set([1])
    unvisited_group = set()
    root = [1]
    for i in range(2, height_type):
        unvisited_group.add(i)
    
    while len(unvisited_group):
        print('경로 :', root)
        print('탐색 그룹 :', visited_group )
        print('미탐색 그룹 :', unvisited_group )
        print('현재 사용 비용 :', answer,'\n')
        
        min_height_diff = 10000
        
        for g1 in list(visited_group):
            for g2 in list(unvisited_group):
                g_diff = get_min_height_diff(land, visit_mat, g1, g2)
                if min_height_diff >=g_diff:
                    next_g = g2
                    min_height_diff = g_diff
        
        answer += min_height_diff
        visited_group.add(next_g)
        unvisited_group.remove(next_g)
    
    return answer
```

```python
from collections import deque

class Group:
    def __init__(self):
        self.connections = []

def solution(land, height):
    N = len(land)

    def is_on(i,j):
        if i>=0 and i<N and j>=0 and j<N:
            return True
        return False
        
    def find_groups():
        visited = [[0 for _ in range(N)] for __ in range(N)]
        groups = [[0 for _ in range(N)] for __ in range(N)]
        vector = [[-1,0],[1,0],[0,-1],[0,1]]
        group = 1
        for i in range(N):
            for j in range(N):
                if visited[i][j]:
                    continue

                cur = land[i][j]
                queue = deque([[i,j, cur]])

                while queue:
                    next_ = queue.popleft()

                    if abs(land[next_[0]][next_[1]] - next_[2])<=height:
                        visited[next_[0]][next_[1]] = 1
                        groups[next_[0]][next_[1]] = group
                        
                        for vr, vc in vector:
                            if is_on(next_[0]+vr, next_[1]+vc) and not visited[next_[0]+vr][next_[1]+vc]:
                                queue.append([next_[0]+vr, next_[1]+vc, land[next_[0]][next_[1]]])
                                
                group+=1
        # for g in groups:
        #     print(g)

    find_groups()

    answer = 0
    return answer
```

```python
from collections import deque
import heapq
import sys
limit_number = 1500000
sys.setrecursionlimit(limit_number)

class Node():
     def __init__(self) -> None:
        self.connections = []

class Tree:
    def __init__(self, group_num) -> None:
        self.nodes = [Node() for i in range(group_num)]
        self.edges = {}


def solution(land, height):
    N = len(land)
    def is_on(i,j):
        if i>=0 and i<N and j>=0 and j<N:
            return True
        return False
    
    
    vector = [[-1,0],[1,0],[0,-1],[0,1]]
    groups = [[0 for _ in range(N)] for __ in range(N)]
    def dfs(i, j, g_num):
        if groups[i][j]:
                return 
        
        groups[i][j]=g_num
        for vr, vc in vector:
            nr, nc = i+vr, j+vc
            if is_on(nr, nc):
                if abs(land[i][j]- land[nr][nc])<=height:
                     dfs(nr, nc, g_num)
                     

    g_num = 1
    for i in range(N):
        for j in range(N):
                if not groups[i][j]:
                    dfs(i,j,g_num)
                    g_num+=1
            
    # for g in groups:
    #     print(g)

    tree = Tree(g_num-1)


    for i in range(N):
        for j in range(N):
            g = groups[i][j]
            for vr, vc in vector:
                nr, nc = i+vr, j+vc
                if is_on(nr, nc):
                    ng = groups[nr][nc]
                    if g != ng:
                         k = (min(g, ng), max(g, ng))
                         if k in tree.edges:
                              tree.edges[k]=min(tree.edges[(min(g, ng), max(g, ng))], abs(land[i][j]-land[nr][nc]))
                         else:
                             tree.edges[(min(g, ng), max(g, ng))] = abs(land[i][j]-land[nr][nc])

    for k, v in tree.edges.items():
         n1, n2 = k
         tree.nodes[n1-1].connections.append([n2-1, v])
         tree.nodes[n2-1].connections.append([n1-1, v])

    # for n in tree.nodes:
    #      print(n.connections)

    hq = [[0,0]] 
    visited = [0 for i in range(g_num-1)]
    answer = 0
    while hq:
        d, cur = heapq.heappop(hq)
        if visited[cur]:
            continue
        answer+= d
        visited[cur] = 1
        for n, dist in tree.nodes[cur].connections:
            heapq.heappush(hq, [dist, n])

        # print(cur, d, hq)

    return answer
        

```
