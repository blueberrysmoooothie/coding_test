```python
import requests
import json
searchText = '경북대학교 정문'
wincId = ''
stid_url = 'https://businfo.daegu.go.kr:8095/dbms_web_api/bs/search?searchText=' + searchText +'&wincId='+wincId
json_txt = requests.get(stid_url).text
stations_json = json.loads(json_txt)['body']
for station in stations_json:
    print(station)
```
