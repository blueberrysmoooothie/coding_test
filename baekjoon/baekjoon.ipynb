```python
input_str = input().split()
word_num = int(input_str[0])
result = 0
for i in range(1,word_num+1):
    word = input_str[i]
    alpha_set = set()
    is_group= True
    before_a = ''
    for a in word:
      if (before_a != a) and a in alpha_set:
        is_group = False
        break
      alpha_set.add(a)
      before_a = a
    if is_group:
      result += 1
        
print(result)
        
        
```

```python
n, m = map(int,input().split())
cards = sorted(map(int,input().split()))
min_dif, result = m, 0
for i in range(n-2):
  for j in range(i+1, n-1):
    for k in range(j+1, n):
      s = cards[i]+cards[j]+cards[k]
      dif = m-s
      if dif >= 0 and min_dif >= dif :
        min_dif = dif 
        result = s
      if dif <0:
        break

print(result)

```

```python
num = int(input())
result = 0
for i in range(1,num):
  n = num - i
  s_nums = list(map(int, list(str(i))))
  if n == sum(s_nums):
    result = i
    break
print(result)

```
